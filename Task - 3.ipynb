Importing relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn import preprocessing
from sklearn.cluster import KMeans
```

Getting the data

```python
df = pd.read_csv('Iris.csv')
df.drop('Id',axis=1,inplace=True)
df.head()
```

```python
df['Species'].value_counts()
```

Getting a quick description of the data

```python
df.info()
```

```python
# Divide the data into inputs and labels

X= df.drop('Species',axis=1)
y = df['Species']
```

```python
X.head()
```

Visualizing features with scatter plot

```python
# Create a scatter plot based on two corresponding features (sepal_length and sepal_width; OR petal_length and petal_width)

sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
plt.scatter(data=X,x='SepalLengthCm',y='SepalWidthCm')
plt.xlabel('Length of sepal')
plt.ylabel('Width of sepal');
```

Standardize the variables

```python
# Scale the variables

X_scaled = preprocessing.scale(X)
X_scaled[:10]
```

The Elbow Method

```python
# Using elbow method to find the optimum num of clusters

# We will us the range b/q w 1 - 10

wcss = []

cl = 10

for i in range(1,cl):
    kmeans = KMeans(i)
    kmeans.fit(X_scaled)
    x = kmeans.inertia_
    wcss.append(x)
wcss      
```

```python
# Ploting the elbow curve
plt.figure(figsize=(10,6))
num_cl = range(1,cl)
plt.plot(num_cl,wcss);
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares');
```

Based on the elbow curve it seems like we can plot our graph with 2 , 3 and 5 no. of clusters .

```python
# 2 cluster

kmeans_2 = KMeans(2)
kmeans_2.fit(X_scaled)

cl_2 = X.copy()

cl_2['pred'] = kmeans_2.fit_predict(X_scaled)

plt.figure(figsize=(10,6))
plt.scatter(cl_2['SepalLengthCm'], cl_2['SepalWidthCm'], c= cl_2['pred'], cmap = 'rainbow');
```

```python
# 3 cluster

kmeans_3 = KMeans(3)
kmeans_3.fit(X_scaled)

cl_3 = X.copy()

cl_3['pred'] = kmeans_3.fit_predict(X_scaled)

plt.figure(figsize=(10,6))
plt.scatter(cl_3['SepalLengthCm'], cl_3['SepalWidthCm'], c= cl_3['pred'], cmap = 'rainbow');
```

```python
# 5 cluster

kmeans_5 = KMeans(5)
kmeans_5.fit(X_scaled)

cl_5 = X.copy()

cl_5['pred'] = kmeans_5.fit_predict(X_scaled)

plt.figure(figsize=(10,6))
plt.scatter(cl_5['SepalLengthCm'], cl_5['SepalWidthCm'], c= cl_5['pred'], cmap = 'rainbow');
```
